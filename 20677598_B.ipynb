Variables:
1.pokemon(dictionary):Storing the information of pokemons e.g. Name:[type,type serial number,HP stat point,attack stat point, defence stat point,special attack stat point,special defence stat point,speed stat point,pokemon ASCII art]

2.type_counter_message(dictionary):Storing the counter message between pokemons e.g. type[damage coefficient towards Fire,damage coefficient towards Grass,damage coefficient towards Water,damage coefficient towards Ground,damage coefficient towards Ice,damage coefficient towards Electric]

3.from_number_to_pokemon(dictionary):Converting number from user to the pokemon they choose e.g.[number:pokemon name]

4.from_number_to_attack_type(dictionary):Converting the number from user to the action they want to take e.g.[number:syntax of the action]

5.round_count(int): Counting the round

6.player_win_count(int): Counting the number of rounds won by user

7.computer_win_count(int): Counting the number of rounds won by computer

8.player_pokemon(str): Storing the pokemon chosen by user

9.computer_pokemon(str): Storing the pokemon chosen randomly by computer

10.player_pokemon_level(int): Storing the pokemon level of user(randomly gernerated from range 1-30)

11.computer_pokemon_level(int): Storing the pokemon level of computer(randomly gernerated from range 1-30)

12.player_HP(int->float): Storing the HP of user's pokemon in current round(HP=HP stat point*50+3*level)

13.computer_HP(int->float): Storing the HP of computer's pokemon in current round(HP=HP stat point*50+3*level)

14.player_action(int): Storing the action made by the user

15.computer_action(int): Storing the random action made by computer

15.computer_damage(int->float): Storing computer's damage to user(randomly attck type's stat point*random number 50-60)

16.player_damage(int->float): Storing user's damage to computer(chosen attck type's stat point*random number 50-60)

17.player_defence_type(int): Soring the corresponding defend type's coefficient(corresponding defence type stat point)

18.computer_defence_type(int): Soring the corresponding defend type's coefficient(corresponding defence type stat point)


Functions:
1.Initiate each round(while loop): 
Randomly initialize HP(randomly gernerate level*3+HP stat point*50)
Randomly initialize damage(chosen attack type's stat point*random integer between 50 and 60)

2.Start the fight(if-elif-else,while loop):
End the game when a life less or equal to zero,who is under 0 first will lose the round. If both are under 0, then it will be a tie game.
Display the result of this round

3.End the fight(if-else)
Go back to the front to star next round if round is not greater than 3.
Compare the number of round of player and computer, who wins more will be the final winner. If both win the same, then it is a tie game.
Display the final result.

4.Choosing pokemon:
Let the user to choose the pokemon(totally 6) they like, ask user to input a number correponding to the pokemon and search the data base to convert the number to pokemon and store it in the variable.

5.Chooing the action(if):
Let the user choose the action they want to take(attack,special attack,run)

6.Determine the counter relationship between pokemons(for loop, if):
Search the corresponding type counter message in the data base and adjust the damage

7.Ajusting the damage with the correponding defence(if-else):
Adjusting the damage to original damage*defence stat point/10 

8.Deside whether will dodge the attck/special attack(if-else):
Using randomly gernerate number(1-100) to compare with 10*speed stat point, if random number is smaller then pokemon will dodge the attack/special attack
